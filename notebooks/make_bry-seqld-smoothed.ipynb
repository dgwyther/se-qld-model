```python
varlist=('u_west','v_west','u_south','v_south','u_east','v_east','u_north','v_north')

def smooth_forcingfield(ds,varlist):
    import scipy as sp
    import scipy.ndimage
    T = ds.temp_time.shape[0]
    for var in varlist:
        print('processing ',var)
        field = ds[var]
        for k in range(T):
            # print('processing time: ',k)
            data = field[k,:,:]


            sigma_y = .510
            sigma_x = .510
            sigma = [sigma_y, sigma_x]

            data_smoothed = sp.ndimage.filters.gaussian_filter(data, sigma, mode='constant')

            field[k] = data_smoothed
        print('done this var')
        ds[var] = field
    return ds
    
ds_new = smooth_forcingfield(frc,varlist)

# temp_backup = temp.copy()
# temp = smooth_initialfield(N,temp)
# salt_backup = salt.copy()
# salt = smooth_initialfield(N,salt)

```
